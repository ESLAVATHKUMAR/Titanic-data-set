```python
from sklearn.datasets import load_iris
```

```python
ld = load_iris()
```

```python
df=pd.DataFrame(ld.data)
```

```python
df['Target']=ld.target
```

```python
df.rename(columns = {0:ld.feature_names[0],
               1:ld.feature_names[1],
               2:ld.feature_names[2],
               3:ld.feature_names[3],},inplace = True)
          
          
```

 # Logistic Regression 
 

```python
X=df.drop('Target',axis=1) # X Variable 
```

```python
y=df['Target']    # y Variable 
```

```python
#splitting
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=2549)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logreg =  LogisticRegression()
```

```python
logreg.fit(X_train,y_train)
```

```python
train_res=logreg.score(X_test,y_test)  # for testing we use 
```

```python
train_res = logreg.score(X_train,y_train)
```

```python
train_res
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
for i in range (1,61):
    kncl =  KNeighborsClassifier(n_neighbors = i)
    kncl.fit(X_train,y_train)
    print(i,kncl.score(X_test,y_test))

```
